# Linear Programming

## Production Planning Problem: an example

A firm produces two products, A and B using three different materials I, II and III. The production of A requires 1 unit of I, 4 unit of II and 3 units of III. The production of B requires 2 unit of I, 4 units of II and 1 unit of III. The profit per unit of A is 5 and the profit per unit of B is 4. The firm has 80 units of I, 180 units of II and 90 units of III available. The firm wants to determine how much of each product to produce in order to maximize the profit.

The following table summarizes the data:

| Product | I | II | III | Profit |
|---------|---|----|-----|--------|
| A       | 1 | 4  | 3   | 5      |
| B       | 2 | 4  | 1   | 4      |
| Available | 80 | 180 | 90 |        |

Let $x_1$ and $x_2$ be the number of units of A and B to produce, respectively. The objective is to maximize the profit, which is given by $5x_1 + 4x_2$. 

The constraints are given by:

1. $x_1 + 2x_2 \leq 80$
2. $4x_1 + 4x_2 \leq 180$
3. $3x_1 + x_2 \leq 90$
4. $x_1 \geq 0$
5. $x_2 \geq 0$

which means that the number of units of A and B to produce must be non-negative and the amount of materials used in the production of A and B must not exceed the available amount of materials.

Finally, the problem can be formulated as a linear programming problem:

$$
\begin{align*}
    \max &\quad 5x_1 + 4x_2 \\
    \text{s.t.} &\quad x_1 + 2x_2 \leq 80 \\
    &\quad 4x_1 + 4x_2 \leq 180 \\
    &\quad 3x_1 + x_2 \leq 90 \\
    &\quad x_1, x_2 \geq 0
\end{align*}
$$


## Graphical representation
Graphical representation and solution of the problem are shown below.

```python
!pip install matplotlib
!pip install numpy

import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0, 100, 500)
x2_1 = (80 - x1) / 2
x2_2 = (180 - 4 * x1) / 4
x2_3 = 90 - 3 * x1

obj_line_1 = (202.5 - 5 * x1) / 4

plt.figure(figsize=(5, 5))
plt.grid()
plt.plot(x1, x2_1, label=r"$x_1 + 2x_2 = 80$")
plt.plot(x1, x2_2, label=r"$4x_1 + 4x_2 = 180$")
plt.plot(x1, x2_3, label=r"$3x_1 + x_2 = 90$")

plt.plot(x1, obj_line_1, label=r"$5x_1 + 4x_2 = 202.5$", linestyle="--", color="gray")
# 
plt.scatter(22.5, 22.5, color="red", s=50, label="Optimal solution", zorder=5)

x_upper = np.minimum(x2_1, x2_2)
x_upper = np.minimum(x_upper, x2_3)
plt.fill_between(x1, 0, x_upper, where=(x_upper >= 0), color="gray", alpha=0.5)
plt.xlabel(r"$x_1$", fontsize=20)
plt.ylabel(r"$x_2$", fontsize=20)
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xticks(np.arange(0, 101, 10), fontsize=12)
plt.yticks(np.arange(0, 101, 10), fontsize=12)
plt.legend()
```
